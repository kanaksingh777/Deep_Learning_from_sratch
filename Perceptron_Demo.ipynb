```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#create a sample dataset
```

```python
num_sample = 400
cgpa = np.random.randint(2,10,num_sample)
iq  = np.random.randint(1,10,num_sample)
```

```python
placed = np.where((cgpa >=7) & (iq > 5),1,0)
```

```python
df = pd.DataFrame({'cgpa':cgpa,'iq':iq,'placed':placed})
```

```python
df.head(10)
```

```python
sns.scatterplot(x = df['cgpa'],y = df['iq'],hue= df['placed'])
```

```python
from sklearn.linear_model import Perceptron
```

```python
X = df.iloc[:,:2]
y = df.iloc[:,-1]
```

```python
p = Perceptron()
```

```python
p.fit(X,y)
```

```python
#w1 and w2
p.coef_
```

```python
#bias
p.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X.values, y.values ,clf = p ,legend =2 )
```
