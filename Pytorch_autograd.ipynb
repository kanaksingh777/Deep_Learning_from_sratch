
##UNDERSTANDING HOW- Pytorch calculcates gradients for backpropogation
--



```python
import torch 
import torchvision
import matplotlib.pyplot as plt
##Requies grad tells Pytorch we want to store the gradients for this tensor 
#we need to do this if we are working with basic pytorch tensors


x = torch.FloatTensor([4])
x.requires_grad = True

w= torch.FloatTensor([2])
w.requires_grad = True

b= torch.FloatTensor([3])
b.requires_grad = True

y = w *x +b 
#compute  gradients via pytorch's autograd
y.backward()


print("Calculate gradients")
print("dy/dx" , x.grad.item())
print("dy/dw" , w.grad.item())
print("dy/db" , b.grad.item())

```
